### Pie Chart

> A Pie Chart can only display one series of data. Pie charts show the size of items (called wedge) in one data series, proportional to the sum of the items. The data points in a pie chart are shown as a percentage of the whole pie.

```python
brands = ['hp','Dell','Apple','Acer','Lenovo']
market_share = [22.5, 21.8, 13.7, 12.3, 11.2]
plt.pie(market_share,labels=brands) # Basic Pie Chart
plt.title('Market Share of Laptop Brands')
plt.show();
```

```python
# Pie Chart with Percentage to x decimal place (%1.xf%%) where x = number of decimal places
plt.pie(market_share, labels=brands, autopct='%1.2f%%') 
plt.title('Market Share of Laptop Brands')
plt.show();
```

```python
plt.pie(market_share, labels=brands, autopct='%1.1f%%', shadow=True) # Pie Chart with Shadow
plt.title('Market Share of Laptop Brands')
plt.show();
```

```python
plt.pie(market_share, labels=brands, autopct='%1.1f%%', shadow=True, startangle=90) # Pie Chart will start from 90 degree
plt.title('Market Share of Laptop Brands')
plt.show();
```

```python
# Explode the slice based on the list (explode should be a list of length same as number of slices) 
plt.pie(market_share, labels=brands,explode=[0.05,0,0,0,0]  , autopct='%1.1f%%',startangle=90,shadow=True) 
plt.title('Market Share of Laptop Brands')
plt.show();
```

```python
# Pie Chart with different radius
fig,(ax0,ax1) = plt.subplots(ncols=2,figsize=(10,4)) # Create a figure and a set of subplots
ax0.pie(market_share, labels=brands,explode=[0.05,0,0,0,0]  , autopct='%1.1f%%',startangle=90,shadow=True,radius=1) 
ax1.pie(market_share, labels=brands,explode=[0.05,0,0,0,0]  , autopct='%1.1f%%',startangle=90,shadow=True,radius=1.5) 
ax0.set_title('Market Share of Laptop Brands')
ax1.set_title('Market Share of Laptop Brands')
plt.show();
```

```python
plt.pie(market_share, labels=brands,explode=[0.05,0,0,0,0]  , autopct='%1.1f%%',startangle=180,shadow=True,counterclock=False)
# plt.legend(bbox_to_anchor=(1,0,0.5,1)) # Legend outside the pie chart
# plt.legend(bbox_to_anchor=(1,1,0,0)) # Legend outside the pie chart
plt.legend(bbox_to_anchor=(1,1)) # Legend outside the pie chart
plt.title('Market Share of Laptop Brands')
plt.show();
```

### Overlapping Pie Chart

```python
brands = ['hp','Dell','Apple','Acer','Lenovo']
market_share = [22.5, 21.8, 13.7, 12.3, 11.2]
avg_prices = [700, 650, 1200, 450, 500]
c = ['red','green','blue','yellow','orange']
plt.pie(market_share,colors=c ,labels=brands,radius=1.2)
plt.pie(avg_prices,colors=c ,radius=0.7,shadow=True,wedgeprops={'edgecolor':'grey'})
plt.show();
```

### Donut Pie Chart

```python
plt.pie(market_share ,labels=brands)
plt.pie([1],colors="white" ,radius=0.7)
plt.show();
```

### Stem Plot

> A Stem Plot is a discrete series plot, ideal for plotting small sets of data. It is similar to a histogram, but it is used to plot discrete rather than continuous data. The data points are plotted as stems extending from a baseline. The stems are labeled with the x-axis values and the leaves are labeled with the y-axis values.

```python
x = ["David","John","Mark","James","Andrew","Scott"]
y = [-2,3,1,-0.5,0,5]
plt.stem(x,y) # Basic Stem Plot
plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Marks of Students')
plt.show();
```

```python
 # Stem Plot with different line and marker format and color and base line 
# plt.stem(x,y,linefmt="--",markerfmt="D") # D = Diamond , -- = Dashed Line
# plt.stem(x,y,linefmt="--",markerfmt="^") # ^ = Triangle , -- = Dashed Line
# plt.stem(x,y,linefmt="--",markerfmt="s") # s = Square , -- = Dashed Line
# plt.stem(x,y,linefmt="g--",markerfmt="cs") # g = green, c = cyan, s = Square , -- = Dashed Line
# plt.stem(x,y,linefmt="g--",markerfmt="cs",basefmt="b") # b=blue, g = green, c = cyan, s = Square , -- = Dashed Line
plt.stem(x,y,linefmt="g--",markerfmt="cs",basefmt="b--") # b=blue, g = green, c = cyan, s = Square , -- = Dashed Line
plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Marks of Students')
plt.show();
```

```python
plt.stem(x,y,bottom=-2.5) # Stem Plot with different bottom value
plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Marks of Students')
plt.show();
```

```python
plt.stem(x,y,label="Marks") # Stem Plot with line label
plt.xlabel('Students')
plt.title('Marks of Students')
plt.legend()
plt.show();
```

```python
plt.stem(x,y,orientation="horizontal") # Horizontal Stem Plot
plt.xlabel('Students')
plt.ylabel('Marks')
plt.title('Marks of Students')
plt.show();
```

### Box Plot

> A Box Plot is a graphical method of displaying the five-number summary. It is a useful way to visualize the distribution of data due to its ability to show the outliers in the data set.

* ***Circles Represent Outliers***
* ***Box Represents 25th to 75th Percentile***
* ***Line in Box Represents Median***
* ***Vertical Line Outside the box called whiskers represents the range of data***

```python
marks = [0,2,11,19,17,0,43,42,44,51,49,44,50,41,40,51,52,57,58,60,55,55,56,69,66,59,34,80,72,66,59,62,54,39,100,99,97,98]
plt.boxplot(marks) # Basic Box Plot
plt.show();
```

```python
plt.boxplot(marks,vert=False) # Horizontal Box Plot
plt.show();
```

```python
plt.boxplot(marks,notch=True) # Box Plot with Notch (Basically a confidence interval around the median)
plt.show();
```

```python
plt.boxplot(marks,widths=0.7) # Box Plot with different width
plt.show();
```

```python
plt.boxplot(marks,whis=0.3) # Box Plot with different whisker length. (Whisker length = 1.5 * IQR)
plt.show();
```

```python
plt.boxplot(marks,labels=["Marks"]) # Box Plot with different labels
plt.show();
```

```python
plt.boxplot(marks,patch_artist=True) # Box Plot with different color
plt.show();
```

```python
plt.boxplot(marks,patch_artist=True,showmeans=True) # Box Plot with mean
plt.show();
```

```python
plt.boxplot(marks,patch_artist=True,sym="*r",showmeans=True) # Box Plot with different outlier symbol
plt.show();
```

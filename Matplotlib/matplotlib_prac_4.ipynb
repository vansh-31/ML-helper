```python
import numpy as np
import matplotlib.pyplot as plt
```

## Subplots on plot

### matplotlib.pyplot.subplots(nrows=1, ncols=1, *, sharex=False, sharey=False, squeeze=True, width_ratios=None, height_ratios=None, subplot_kw=None, gridspec_kw=None, **fig_kw)

Create a figure and a set of subplots.
> plt.subplots returns a figure and a tuple containing axes objects equal to nrows*ncols. Each axes object is accessible by its index.

```python
# Blanck Axes on Figure
fig, ax = plt.subplots()
```

```python
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 4, 2, 3]);
plt.show()
```

```python
fig, (ax0,ax1) = plt.subplots(nrows=2)
ax0.plot([1, 2, 3, 4], [1, 4, 2, 3])
ax1.plot([1, 2, 3, 4], [4, 1, 1, 4])
plt.show()
```

```python
fig, (ax0,ax1) = plt.subplots(ncols=2,figsize=(12,4))
ax0.plot([1, 2, 3, 4], [1, 4, 2, 3])
ax1.plot([1, 2, 3, 4], [4, 1, 1, 4])
plt.show()
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(12,4))
ax[0,0].plot([1, 2, 3, 4], [1, 4, 2, 3])
ax[0,1].plot([1, 2, 3, 4], [4, 1, 1, 4])
ax[1,0].plot([1, 2, 3, 4], [1, 4, 4, 2])
ax[1,1].plot([1, 2, 3, 4], [2, 2, 2, 2])
```

### Sharing axes

```python
maths = [90, 55, 40, 65, 70, 80, 75, 85, 90, 100]
science = [80, 60, 70, 80, 90, 85, 95, 100, 90, 80]
english = [80, 90, 95, 85, 80, 85, 90, 95, 100, 100]
students = ["Bob","Alice","John","Ann","Mike","Tom","Jane","Kate","Alex","Mary"]
fig,(ax0,ax1,ax2) = plt.subplots(ncols=3,figsize=(18,4),sharey=True) 
ax0.bar(students,maths)
ax0.set_xlabel("Maths")
ax0.set_ylabel("Scores")
ax1.bar(students,science)
ax1.set_xlabel("Science")
ax2.bar(students,english)
ax2.set_xlabel("English")
fig.suptitle("Students' Scores")
plt.show();
```

## Text and annotation in Matplotlib.pyplot

* ***text :*** Add text at an arbitrary location of the Axes.
* ***annotate :*** Annotate an annotation, with an optional arrow, at an arbitrary location in the Axes.
* ***xlabel :*** Set the label for the x-axis.
* ***ylabel :*** Set the label for the y-axis.
* ***title :*** Set a title for the Axes.

```python
x = [1,2,3,4,5,6,7]
y = [4,2,6,9,5,7,4]
plt.plot(x,y)

plt.title("Line Plot",fontdict={"fontsize":16}) # Set the title of the plot with font size 16
plt.xlabel("Days",fontdict={"fontsize":14}) # Set the x-axis label with font size 14
plt.ylabel("Sales",fontdict={"fontsize":14}) # Set the x-axis label with font size 14
plt.ylim(0,12) # Set the y-axis limits

# Add text to the plot at the specified x and y coordinates
# # plt.text(3.5,9.6,"Highest",fontsize=12,color="g")
# # plt.text(1.6,1.2,"Lowest",fontsize=12,color="r")
# Add text to the plot at the specified x and y coordinates in a box with light grey background
plt.text(3.5,9.6,"Highest",fontsize=12,color="g",bbox={"facecolor":"lightgrey"})
plt.text(1.6,1.2,"Lowest",fontsize=12,color="r",bbox={"facecolor":"lightgrey"})

plt.annotate("Point",xy=(4,9),xytext=(5,10),arrowprops={"color":"b"})
plt.annotate("Point",xy=(2,2),xytext=(3,2),arrowprops={"color":"b"})

plt.show();
```

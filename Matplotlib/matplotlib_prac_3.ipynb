```python
import numpy as np
import matplotlib.pyplot as plt
```

### Stack Plot

> A Stack Plot is a plot that shows the whole data set with easy visualization of how each part makes up the whole. Each constituent of the stack plot is stacked on top of each other. It shows the part makeup of the unit, as well as the whole unit.

```python
x = [1,2,3,4,5]
area = [10,20,10,40,30]
plt.stackplot(x, area) # Basic stackplot
plt.show();
```

```python
x = [1,2,3,4,5]
area1 = [10,20,20,20,30]
area2 = [20,10,20,10,30]
area3 = [30,50,10,30,20]
labels = ['area1','area2','area3']
plt.stackplot(x, area1,area2,area2,labels=labels) # Multiple stackplot
plt.legend(loc='upper left')
plt.show();
```

```python
plt.stackplot(x, area1,area2,area2,labels=labels,baseline='sym') # Symmetric stackplot (default is 'zero')
plt.legend(loc='upper left')
plt.show();
```

### Step Plot

> A Step Plot is a plot that shows the data as a series of steps. It is similar to a line plot, but instead of drawing straight lines connecting the data points, a step plot draws lines that connect the center of the data points.

```python
x = [1,2,3,4,5,6]
y = [8,10,12,11,7,13]
plt.step(x,y) # Basic step plot
plt.show();
```

```python
x = [1,2,3,4,5,6]
y = [8,10,12,11,7,13]
# Step plot where parameter defines where the step is taken (default is 'pre') (Basically where the step starts from)
plt.step(x,y,where='pre',color='c') 
plt.step(x,y,where='mid',color='orange')
plt.step(x,y,where='post',color='yellow')
plt.grid(True) # Add grid layout
plt.show();
```

```python
# Add markers to the step plot (marker represents where is data point)
plt.step(x,y,where='pre',color='c',marker='o')
plt.step(x,y,where='mid',color='orange')
plt.step(x,y,where='post',color='yellow')
plt.grid(True) # Add grid layout
plt.show();
```

```python
# Add labels to the step plot
plt.step(x,y,where='pre',color='c',label="python")
plt.step(x,y,where='mid',color='orange',label="java")
plt.step(x,y,where='post',color='yellow',label="c++")
plt.grid(True) # Add grid layout
# plt.legend(loc="upper left") # Add legend
plt.legend(bbox_to_anchor=(1,1)) # Add legend outside the plot
plt.show();
```

### Fill Between Plot

> A Fill Between Plot is a plot that is used to highlight the region between two line plots. It is used to show the difference between the two plots.

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])
plt.plot(x,y) # Basic plot
plt.fill_between(x,y) # Fill the area between the line and the x-axis
plt.show();
```

```python
plt.plot(x,y)
plt.fill_between(x=[2,4],y1=2,y2=4)
plt.show();
```

```python
plt.plot(x,y)
plt.fill_between(x,y,where=((x>=2)&(y<=6)),color='orange') # Fill the area between the line and the x-axis where the condition is true
plt.show();
```

```python
sin = np.sin(np.arange(0,10,0.1))
cos = np.cos(np.arange(0,10,0.1))
plt.plot(sin,color='c',label='sin')
plt.plot(cos,color='orange',label='cos')
plt.fill_between(x=np.arange(0,100,1),y1=sin,y2=cos,color='r',alpha=0.4) # Fill the area between sin curve and cos curve
plt.xlabel('input data')
plt.ylabel("Trignometric values")
plt.legend(loc='upper left')
plt.show();
```

```python
sin = np.sin(np.arange(0,10,0.1))
cos = np.cos(np.arange(0,10,0.1))
plt.plot(sin,color='c',label='sin')
plt.plot(cos,color='orange',label='cos')
plt.fill_between(x=np.arange(0,100,1),y1=sin,y2=cos,where=((sin>=cos)),color='r',alpha=0.4) # Fill the area with red color where sin is greater than or equal to cos
plt.fill_between(x=np.arange(0,100,1),y1=sin,y2=cos,where=((sin<cos)),color='b',alpha=0.4) # Fill the area with blue color where cos is greater than sin
plt.xlabel('input data')
plt.ylabel("Trignometric values")
# plt.legend(loc='upper left') # Add a legend
plt.legend(bbox_to_anchor=(1.2,1)) # Add a legend outside the plot
plt.show();
```

```python
import pandas as pd
df = pd.read_excel("stocks.xlsx",header=[0,1],index_col=0)
df
```

```python
df.stack()
```

```python
df.stack(level=0)
```

```python
df_stacked=df.stack()
df_stacked
```

```python
df_stacked.unstack()
```

### Strip Plot

A strip plot is a scatter plot where one variable is categorical. It is used to visualize the distribution of data points of one variable.

```python
df = sns.load_dataset('tips')
df.head()
```

```python
# sns.stripplot(data=df,x='day',y='total_bill') # Basic stripplot
# sns.stripplot(data=df,x='day',y='total_bill',hue="sex") # hue is used to plot different colors for different values of a feature
# sns.stripplot(data=df,x='day',y='total_bill',hue="sex",jitter=2) # jitter is used to add noise to the data
sns.stripplot(y=df.total_bill) # Single feature stripplot
plt.show();
```

### Swarm Plot

A swarm plot is a scatter plot where one variable is categorical. It is used to visualize the distribution of data points of one variable. It is similar to a strip plot but the points are adjusted so that they do not overlap.

```python
# sns.swarmplot(data=df,x='day',y='total_bill') # Basic stripplot
sns.swarmplot(data=df,x='day',y='total_bill',hue="sex") # hue is used to plot different colors for different values of a feature
# sns.swarmplot(y=df.total_bill) # Single feature stripplot
plt.show();
```

### ECDF Plot

ECDF stands for Empirical Cumulative Distribution Function. It is used to visualize the distribution of a variable. It is a step function that jumps up by 1/n at each of the n data points. It is useful to visualize the distribution of a variable.

```python
df = sns.load_dataset('tips')
df
```

```python
# sns.ecdfplot(data=df,x='total_bill'); # Basic ecdfplot
# sns.ecdfplot(data=df,x='total_bill',hue="sex"); # Basic ecdfplot
# plt.vlines(x=20,ymin=0,ymax=1,linestyles='dashed') # vlines is used to draw vertical lines
# plt.hlines( y=0.55,xmin=0,xmax=50,linestyles="--" ) # hlines is used to draw horizontal lines
# stat can take values: {‘proportion’, ‘count’, ‘frequency’, ‘density’, ‘probability’} (default: ‘proportion’)
sns.ecdfplot(data=df,x='total_bill',stat='count'); # stat is used to change the y-axis values (count=number of values)
```

### Joint Plot

A joint plot is a plot where the relationship between two variables is visualized along with the distribution of each variable. It is used to visualize the relationship between two variables and the distribution of each variable.

By default it is a scatter plot with a histogram on each axis. It can also be a regression plot with a histogram on each axis.

Joint Plot takes kind arguments which can take these values : scatter, reg, resid, kde, hex. (default = scatter)

```python
sns.jointplot(data=df,x='total_bill',y='tip'); # Basic jointplot
# sns.jointplot(data=df,x='total_bill',y='tip',kind="reg");
# sns.jointplot(data=df,x='total_bill',y='tip',kind="resid");
# sns.jointplot(data=df,x='total_bill',y='tip',kind="kde");
# sns.jointplot(data=df,x='total_bill',y='tip',kind="hex");
```

### Box Plot

A box plot is a plot that shows the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable. The box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be “outliers” using a method that is a function of the inter-quartile range.

```python
df = sns.load_dataset('tips')
df
```

```python
# sns.boxplot(data=df,x='day',y='total_bill') # Basic boxplot
# sns.boxplot(data=df,x='day',y='total_bill',hue="time") # hue is used to plot different colors for different values of a feature
# sns.boxplot(data=df,x='day',y='total_bill',hue="time",palette="Set3") # palette is used to change the color scheme
sns.boxplot(data=df,x='day',y='total_bill',hue="time",palette="Set3",showmeans=True) # showmeans is used to show the mean of the data
```

### Boxen Plot

A boxen plot is similar to a box plot in plotting a nonparametric representation of a distribution in which all features correspond to actual observations. By plotting more quantiles, it provides more information about the shape of the distribution, particularly in the tails.

```python
sns.boxenplot(data=df,x='day',y='total_bill') # Basic boxenplot
# sns.boxenplot(data=df,x='day',y='total_bill',hue="time") # hue is used to plot different colors for different values of a feature
# sns.boxenplot(data=df,x='day',y='total_bill',hue="time",palette="Set3") # palette is used to change the color scheme
```

### Cat Plot

A cat plot is a plot where one of the variables is categorical. It provides a high-level interface for drawing categorical plots.

It takes a kind parameter to adjust the type of plot to draw.
kind can take these values: “point”, “bar”, “strip”, “swarm”, “box”, “violin”, or “boxen”. (default: “strip”)

```python
df = sns.load_dataset('tips')
df
```

```python
sns.catplot(data=df,x='day',y='total_bill') # Basic catplot
# kind Options are 'strip', 'swarm', 'box', 'boxen', 'violin', 'bar', 'count', and 'point'.
# sns.catplot(data=df,x='day',y='total_bill',kind="strip") # kind is used to select the type of plot to be plotted (strip=strip plot)
# sns.catplot(data=df,x='day',y='total_bill',kind="swarm") # Swarm plot is similar to strip plot but the points are adjusted so that they don't overlap
# sns.catplot(data=df,x='day',y='total_bill',kind="box") 
# sns.catplot(data=df,x='day',y='total_bill',kind="boxen") 
# sns.catplot(data=df,x='day',y="total_bill",kind="violin") 
# sns.catplot(data=df,x='day',y='total_bill',kind="bar") 
# sns.catplot(data=df,x='day',kind="count") 
# sns.catplot(data=df,x='day',y='total_bill',kind="point") # Point plot is similar to strip plot but the points are joined by a line (joined point are used to show the mean of the data)
plt.show();
```

## Facet Grid

FacetGrid is a multi-plot grid for plotting conditional relationships. It is used to draw plots with multiple Axes where each Axes shows the same relationship conditioned on different levels of some variable. It is possible to condition on up to three variables by assigning variables to the rows and columns of the grid and using different colors for the plot elements.

```python
df = sns.load_dataset('tips')
df
```

```python
graph = sns.FacetGrid(df,col="day") # FacetGrid is used to plot multiple plots side by side
graph.map(sns.histplot,"total_bill") # histplot is used to plot a histogram
```

### Heatmap

A heatmap is a graphical representation of data where the individual values contained in a matrix are represented as colors. A heatmap is very useful in visualizing the concentration of values between two dimensions of a matrix. The data used in a heatmap is often represented as a matrix and is presented in the form of a table.

* ***Heatmap is one of the best way to plot the correlation matrix***
* ***It only works with numerical values***

```python
data = np.linspace(1,10,25).reshape(5,5)
data
```

```python
sns.heatmap(data); # Basic heatmap
```

### Count Plot

A countplot is kind of likea histogram or a bar graph for some categorical area. It simply shows the number of occurrences of an item based on a certain type of category. It is one of the most basic plots provided by the seaborn library.

***Difference between countplot and barplot***

* ***Countplot shows the counts of observations in whereas barplot shows the mean of observations in each category.***
* ***Countplot can be thought of as a histogram across a categorical, instead of quantitative, variable.***

```python
df = sns.load_dataset("tips")
df
```

```python
sns.countplot(data=df,x="day"); # Basic countplot
```

> Count Plot vs Bar Plot

```python
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,5)) # We can create multiple plots in a single figure
# sns.countplot(data=df,x="day",y="total_bill",ax=ax1); # It will throw an error because countplot can only take x or y
sns.countplot(data=df,x="day",ax=ax1); # It will throw an error because countplot can only take x or y
sns.barplot(data=df,x="day",y="total_bill",ax=ax2); # Basic barplot
```

***In the above cell we can observe that the countplot is showing the count of each category whereas the barplot is showing the mean of each category and in barplot we also have to pass the y axis values which is not required in countplot since it is already showing the count of each category and barplots shows the mean of each category of the passed y axis values.***

```python
# sns.countplot(data=df,x="day") # Basic countplot
# sns.countplot(data=df,x="day",hue="sex") # hue is used to show the data in different colors based on the column
# sns.countplot(data=df,x="day",hue="sex",palette="viridis") # Palette is used to change the color of the hue
# sns.countplot(data=df,x="day",hue="sex",palette="viridis") # Palette is used to change the color of the hue
sns.countplot(data=df,x="day",hue="sex",palette="viridis",alpha=0.6) # Alpha is used to change the transparency of the data
```
